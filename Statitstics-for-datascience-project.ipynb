## T-test :

### Hypothesis :

##### H0: Variances are equals v1=v2
##### H1: Variances are unequals

##### Test for equality variance

```python
scipy.stats.levene(boston_df[boston_df['CHAS']==0]['MEDV'],
                  boston_df[boston_df['CHAS']==1]['MEDV'], center='mean')
```

#### Conclusion: We reject the null hypothesis for the 5% threshold, the variances are unequals .

### Hypothesis :

##### H0: The mean of median value of houses bounded by the Charles river is equal of those who's not bounded by the charles river
##### H1: The mean of median value of houses bounded by the Charles river is different of those who's not bounded by the charles river

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS']==0]['MEDV'],
                  boston_df[boston_df['CHAS']==1]['MEDV'], equal_var = False)
```

#### Conclusion: the pvalue is lower than 5%, we reject the null hypothesis, so the mean of median value of houses bounded by the Charles river is different of those who's not bounded by the charles river.

## ANOVA

### Hypothesis :


##### H0: variances are equals between the three groups v1=v2=v3
##### H1: There is at less one variance different to others 

#### Test for equality variance :

```python
scipy.stats.levene(boston_df[boston_df['CAGE'] == '35 and younger']['MEDV'],
                   boston_df[boston_df['CAGE'] == 'between 35 and 70 years']['MEDV'], 
                   boston_df[boston_df['CAGE'] == '70 years and older']['MEDV'], 
                   center='mean')
# since the p-value is less than 0.05, the variance are not equal, for the purposes of this exercise, we will move along
```

#### Conclusion: we fail to reject the null hypothesis for the 5% threshold, so the variances are equals.

### Hypothesis :

##### H0: mean of median value of houses are equals between the three groups.
##### H1: There is at less one mean different to others 

```python
thirtyfive_lower = boston_df[boston_df['CAGE'] == '35 and younger']['MEDV']
thirtyfive_seventy = boston_df[boston_df['CAGE'] == 'between 35 and 70 years']['MEDV']
seventy_older = boston_df[boston_df['CAGE'] == '70 years and older']['MEDV']
```

```python
scipy.stats.f_oneway(thirtyfive_lower, thirtyfive_seventy, seventy_older)
```

#### Conclusion: we reject the null hypothesis for the 5% threshold, so the mean of median value of houses are unequals between the three groups.

```python
[boston_df[boston_df['CAGE'] == '35 and younger']['MEDV'].mean(),
boston_df[boston_df['CAGE'] == 'between 35 and 70 years']['MEDV'].mean(),
boston_df[boston_df['CAGE'] == '70 years and older']['MEDV'].mean()]
```
